#1.  What is a Common Table Expression (CTE), and how does it improve SQL query readability?

solution:- A CTE (Common Table Expression) is a temporary result set that you can define using with keyword and use within a query


syntax:-    WITH cte_name AS

(
    SELECT column1, column2
    
    
    FROM table_name
)


SELECT * FROM cte_name;




----  How it improves readability:-



Breaks complex queries into smaller parts


Makes SQL easier to understand


Avoids repeated subqueries.



#2. Why are some views updatable while others are read-only? Explain with an example.

solution:-  A view is updatable when:

It is based on a single table


No aggregate functions (SUM, AVG, etc.)


No GROUP BY or DISTINCT

# A view is read-only when:

It uses JOIN


Contains aggregation


Uses GROUP by

#Example:

✔ Updatable view:

CREATE VIEW simple_view AS


SELECT ProductID, ProductName, Price


FROM Products;


# Read-only view:

CREATE VIEW summary_view AS


SELECT Category, COUNT(*)


FROM Products


GROUP BY Category;

Maintain data integrity
