9. What is the Difference Between a Primary Key and a Foreign Key?

-> solution:-     Primary Key


A Primary Key is a column (or set of columns) that uniquely identifies each row in a table


It cannot have duplicate values


It cannot be NULL


Each table can have only one primary key

Foreign Key


A Foreign Key is a column that refers to the Primary Key of another table


It is used to create a relationship between two tables


It can have duplicate values


It can have NULL values (in most cases)


A table can have multiple foreign keys


10.  What Are Constraints in SQL and Why Are They Used?

-> Solution:-  Constraints are rules applied to table columns in SQL to control the type of data that can be stored.

They help ensure that the data in the database is accurate, valid, and reliable.

Constraints are used to:

✔ Maintain data accuracy


✔ Ensure data consistency


✔ Prevent invalid data entry


✔ Enforce rules automatically


✔ Maintain relationships between tables
