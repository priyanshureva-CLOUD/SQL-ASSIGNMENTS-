#Question 3 : What is duplicate data? Explain three causes in ETL pipelines.

#solution:-
Duplicate data in ETL (Extract, Transform, Load) refers to the situation where

 the same data record is stored, processed, or loaded multiple times within a

 data pipeline or database, even though only one unique version of that record

 is required. These duplicate records may be completely identical or may

  contain slightly different values while still representing the same
  
  real-world entity. Duplicate data is usually created because of repeated
  
  extraction from source systems, improper joins during transformation,
  
   multiple file imports, rerunning ETL jobs, missing primary keys, or
   
   incorrect loading logic.


Three causes of duplicate data in ETL pipelines:


1. Multiple Data Sources: Same data comes from different systems.


2. Repeated ETL Loads: ETL jobs may run multiple times accidentally.


3. Lack of Primary Keys: Without unique identifiers, duplicate records cannot

be detected properly.


#Question 4 : Differentiate between exact, partial, and fuzzy duplicates.

#solution:-
#Exact Duplicates:
Records are completely identical.


#Example:
Same Customer_ID, Product_ID, Txn_Date, and Amount.


#Partial Duplicates:
Some fields are same while others are slightly different.


#Example:
Same customer and product but different quantity


#Fuzzy Duplicates:
Records look similar but are not exactly same due to spelling

 or formatting differences.


#Example:
“Rahul Mehta” and “Rahul Mehata”.


#Question 5 : Why should data validation be performed during transformation rather than after loading?

#solution:-
Data validation should be performed during transformation because invalid data

 can be corrected before entering the target database.

 This reduces errors, avoids bad reports, and improves system performance.


 #Benefits:


• Prevents loading of incorrect records


• Saves storage space


• Improves dashboard accuracy


• Reduces rework after loading

Referential integrity ensures that every Customer_ID in Sales_Transactions

exists in Customers_Master.

```python
#7    SELECT Customer_ID,
Product_ID,
Txn_Date,
Txn_Amount,
COUNT(*) AS Duplicate_Count
FROM Sales_Transactions
GROUP BY Customer_ID, Product_ID, Txn_Date, Txn_Amount
HAVING COUNT(*) > 1;


#8   SELECT s.Customer_ID
FROM Sales_Transactions s
LEFT JOIN Customers_Master c
ON s.Customer_ID = c.CustomerID
WHERE c.CustomerID IS NULL;


```
