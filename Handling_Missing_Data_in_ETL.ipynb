#3.  Explain the difference between:

Listwise deletion

Column deletion

Also mention one scenario where each is appropriate.

#SOLUTION:-  Listwise Deletion (Row-wise deletion)


#What it does:

Removes entire rows (records) from a dataset if any value in that row is missing.


#Impact:

You lose complete observations, even if only one variable is missing.

#Example:


If a dataset has 100 rows and 10 rows have at least one missing value, those 10 rows are completely removed → leaving 90 rows.

#When it’s appropriate:


When the amount of missing data is small and randomly distributed (often called Missing Completely at Random in Missing Data Theory).


When preserving relationships between variables is more important than keeping all cases.

#Scenario:


A survey dataset where only 2–3% of responses have missing answers scattered randomly. Removing those few rows won’t bias results much.


#Column Deletion (Variable deletion)


#What it does:

Removes entire columns (variables/features) that contain missing values.
Impact: You lose variables (features), but keep all observations (rows).

#Example:


If one column has 60% missing data, deleting that column keeps all rows but removes that feature entirely.


#When it’s appropriate:

When a variable has too many missing values to be useful.


When the variable is not critical for analysis or modeling.

#Scenario:


In a dataset of patient records, a column like “middle name” or “secondary contact number” is missing for most entries and isn’t important → safe to drop the column.


#4.  Why is median imputation preferred over mean imputation for skewed data such as income?

#SOLUTION:-

Median imputation is usually preferred over mean imputation for

skewed data like income because of how each statistic reacts to extreme values.

In a skewed distribution, especially something like income, a small number of

very high values (outliers) pull the mean upward. This is a direct consequence

of how the Arithmetic Mean is calculated—it uses all values equally, so extreme

ones have a strong influence. If you use mean imputation, you end up filling

missing values with a number that may be much higher than what’s typical for

most individuals.




The median, on the other hand, is the middle value in the ordered data and is

much more robust to outliers. This property is tied to the Median being a

positional measure rather than a magnitude-based average. So even if a few

incomes are extremely large, the median still reflects the “typical” person

much better.


#Why this matters in practice:

Mean imputation in skewed data can overestimate missing values, distorting the dataset.


Median imputation preserves the central tendency of the majority of observations.


It leads to less biased analyses, especially when modeling or summarizing income-like variables.


#5  What is forward fill and in what type of dataset is it most useful?

#SOLUTION:-
Forward fill (also called ffill) is a method to handle missing values where

each missing entry is replaced with the last known (previous) value in the

dataset.

#Best-suited dataset types:


Time-series data


Stock prices


Daily sales


Weather data



Ordered/sequence data

#6.  Why should flagging missing values be done before imputation in an ETL workflow?

#SOLUTION:-

Flagging missing values before imputation is important because imputation hides

 the fact that data was originally missing.

 #Why flag first?


1. Preserve original information


A flag column (e.g., Income_Missing_Flag) keeps track of missingness


After imputation, everything looks “filled” → but flag tells the truth


2. Better analysis & insights


Missing data itself can carry meaning


Example: People not reporting income might belong to a specific group


3. Improves machine learning models


Models can use the flag as an extra feature


Sometimes missingness is predictive


4. Avoids hidden bias


Without flags, imputed values may be treated as real → misleading results
